# Word Embedding Analysis
This notebook will help you analyze your text data for word embedding training.

```python
import pandas as pd

# Read the CSV file with Content column
file = "news_dataset.csv"
df = pd.read_csv(file, usecols=['Content'])

# Write each content to a text file, one per line
output_file = "news_content.txt"
with open(output_file, 'w', encoding='utf-8') as f:
    for content in df['Content']:
        if pd.notna(content):  # Check if content is not NaN
            f.write(str(content) + '\n')

print(f"Successfully wrote {len(df)} content entries to {output_file}")
```

```python
import pandas as pd
import numpy as np
import re
file = "news_text.txt"

with open(file, 'r', encoding='utf-8') as f:
    lines = f.readlines()

# Remove leading and trailing whitespace from each line
lines = [line.strip() for line in lines]
lines = [line.lower() for line in lines]

modified_lines = []

for line in lines:
    line = re.sub(r'[0-9]', '', line) 
    line = re.sub(r"https?://\S+|www\.\S+", "<url>", line) 
    line = re.sub(r"[\"#\$%&\(\)\*\+,/:;<=>@\[\\\]\^_`{\|}~]", "", line)
    if (' - ' in line):
         line = line.split(' - ', 1)[1]
    if (' -- ' in line):
         line = line.split(' -- ', 1)[1]
    if('  – ' in line):
         line = re.split('  – ',line)[1]
    
    line = line.replace('kompas.com', '')
    line = line.replace('kompastekno', '')
    line = line.replace('republika.co.id', '')
    line = line.replace('tribunnews.com', '')
    line = str(line)
    line = re.sub(r'\.{2,}', '.', line)  # Replace .. and more with .
    line = line.replace('"', '')          # Remove "
    line = line.replace('-', ' ')         # Replace - with space
    line = line.strip()
    modified_lines.append(line)

# Write the processed lines back to the file
with open(file, 'w', encoding='utf-8') as f:
    f.write('\n'.join(modified_lines))

print(f"Processed {len(modified_lines)} lines in {file}")
```
