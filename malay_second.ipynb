```python
# Preprocessing for shahbudin
import json
import re

input_file = "dirty_malaysian_text/alhijrahnews-articles.jsonl"
text_only = []
def analyse():
    # Counter to limit output to first 5 lines
    count = 0
    with open(input_file, 'r', encoding='utf-8') as infile:
        for line in infile:
            try:
                # Parse JSON line
                
                data = json.loads(line)
                data = re.sub('\n', '', data)
                data = re.sub('\t', '', data)
                data = re.sub('\r', '', data)
                data = re.sub('\s+', ' ', data)
                data = re.sub('\u00a0','',data)
                data = re.sub('votesArticle RatingKongsikan artikel ini', '', data)
                data = re.sub('\[email\u00a0protected\]', '', data)
                
                date_patterns = [
                    r'\b\d{1,2}[/-]\d{1,2}[/-]\d{2,4}\b',  # DD/MM/YYYY or DD-MM-YYYY
                    r'\b\d{4}[/-]\d{1,2}[/-]\d{1,2}\b',    # YYYY/MM/DD or YYYY-MM-DD
                    r'\b(?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{1,2},? \d{4}\b',  # Month name dates
                    r'\b\d{1,2} (?:Jan|Feb|Mar|Apr|May|Jun|Jul|Aug|Sep|Oct|Nov|Dec)[a-z]* \d{4}\b'     # DD Month YYYY
                ]
                for pattern in date_patterns:
                    data = re.sub(pattern, '', data)
                data = re.sub(r'[^\w\s]', '', data)
                data = re.sub(r'\d', '', data)
                data = re.sub(' SH ','',data)
                # text_only.append(data[0])
                print(f"--- Line {count+1} ---")
                print(data)
                print(f"--- Line ends ---")
                print("\n")
                count += 1
                text_only.append(data)
                
                # Stop after 5 lines to avoid flooding output
                if count >= 50:
                    print("Showing first 5 lines only...")
                    break
                    
            except json.JSONDecodeError:
                print(f"Error decoding line {count+1}")
                # Print the problematic line for debugging
                print(f"Problematic content: {line[:100]}...")
                count += 1

# Run the analysis
analyse()
```
